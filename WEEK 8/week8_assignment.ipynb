**Implementing OOP Concepts**

```python
class SalesDataProcessor:
    def __init__(self,file_path):
        self.df=self.load_data(file_path)
    def load_data(self,file_path):
        return pd.read_csv(file_path)
    def get_total_sales(self):
        return self.df["TotalPrice"].sum()
    def get_unique_products(self):
        return self.df["Product"].unique().tolist()
    def get_sales_by_category(self):
        return self.df.groupby("Category")["TotalPrice"].sum().to_dict()
    def get_top_selling_product(self):
        return self.df.groupby("Product")["TotalPrice"].sum().idxmax()
obj=SalesDataProcessor('sales_data_india.csv')
print("Total Sales:", obj.get_total_sales())
print("Unique Products:",obj.get_unique_products())
print("Sales by Category:",obj.get_sales_by_category())
print("Top Selling Product:",obj .get_top_selling_product())
```

```python
class CustomerSalesProcessor(SalesDataProcessor):
    def __init__(self,file_path):
        super().__init__(file_path)
    def get_total_sales_by_customer_sales(self, CustomerID):
        return self.df.loc[self.df["CustomerID"] == CustomerID, "TotalPrice"].sum()
    def get_frequent_customers(self,n):
        return self.df.groupby("CustomerID")["TotalPrice"].sum().sort_values(ascending=False).head(n).index.to_frame(index=False).set_index(pd.RangeIndex(1, n+1))
    def get_sales_by_city(self):
        return self.df.groupby("City")["TotalPrice"].sum().to_dict()
obj2 = CustomerSalesProcessor('sales_data_india.csv')
print("Total Sales by Customer C001:", obj2.get_total_sales_by_customer_sales("C001"))
print("customers who made the most purchases:\n", obj2.get_frequent_customers(5))
print("total sales per city:\n", obj2.get_sales_by_city())
```

**Data Visualization**

```python
import matplotlib.pyplot as plt

def plot_sales_by_category(sales_dict):
    categories = list(sales_dict.keys())
    totals = list(sales_dict.values())

    plt.figure(figsize=(8, 5))
    plt.bar(categories, totals,width = 0.3, color='blue')
    plt.xlabel('Category')
    plt.ylabel('Total Sales')
    plt.title('Total Sales by Category')
    plt.xticks(rotation=45)
    plt.grid(False)
    plt.tight_layout()
    plt.show()
obj = SalesDataProcessor('sales_data_india.csv')
category_sales = obj.get_sales_by_category()
plot_sales_by_category(category_sales)
```
