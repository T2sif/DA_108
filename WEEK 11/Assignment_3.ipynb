# **Image Processing (Grayscale & Black-White Conversion)**

# Objective: Convert images into grayscale and binary (black and white) using OpenCV

**Tasks:**
1. Load a color image using cv2.imread().
2. Convert the image to grayscale using cv2.cvtColor(image, cv2.COLOR_BGR2GRAY).
3. Convert the grayscale image to a binary black-and-white image using thresholding
(cv2.threshold()).
4. Display the original, grayscale, and binary images using matplotlib.pyplot.imshow().
5. Save the grayscale and black-and-white images using cv2.imwrite

```python
import cv2
import matplotlib.pyplot as plt
```

**load image**

```python
img=cv2.imread('Phase_1_HD.png')
```

**Convert BGR to RGB**

```python
img_rgb=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
```

**Convert the image to grayscale**

```python
img_gray=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
```

**Convert the grayscale image to a binary black-and-white image**

```python
img_binary=cv2.threshold(img_gray,127,255,cv2.THRESH_BINARY)[1]
```

**Display the original, grayscale, and binary images**

```python
#original
plt.figure(figsize=(10, 4))
plt.subplot(1, 3, 1)
plt.imshow(img_rgb)
plt.title("Original Image")
plt.axis("off")

#grayscale
plt.subplot(1, 3, 2)
plt.imshow(img_gray, cmap="gray")
plt.title("Grayscale Image")
plt.axis("off")

#binary
plt.subplot(1, 3, 3)
plt.imshow(img_binary, cmap="gray")
plt.title("Binary Image")
plt.axis("off")
plt.savefig("new_img.png", bbox_inches='tight', dpi=300)
plt.show()
```
